```python
## Lower Casing
data["Services"]=data["Services"].str.lower()
data["Menu"]=data["Menu"].str.lower()


## Removal of Puctuations
import string
PUNCT_TO_REMOVE = string.punctuation
def remove_punctuation(text):
    """custom function to remove the punctuation"""
    return text.translate(str.maketrans('', '', PUNCT_TO_REMOVE))

#data["Services"] = data["Services"].apply(lambda text: remove_punctuation(text))
data["Menu"] = data["Menu"].apply(lambda text: remove_punctuation(text))
data["Services"]=data["Services"].apply(lambda text: remove_punctuation(text))

## Removal of Stopwords
from nltk.corpus import stopwords
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])

#data["Services"] = data["Services"].apply(lambda text: remove_stopwords(text))
data["Menu"] = data["Menu"].apply(lambda text: remove_stopwords(text))
data["Services"]=data["Services"].apply(lambda text: remove_stopwords(text))

## Removal of URLS
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

#data["Services"] = data["Services"].apply(lambda text: remove_urls(text))
data["Menu"] = data["Menu"].apply(lambda text: remove_urls(text))
data["Services"]=data["Services"].apply(lambda text: remove_urls(text))


data[['Menu', 'Services']].sample(5)
#data["Services"].sample(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
Services_list = list(data['Services'].unique())
def get_top_words(column, top_nu_of_words, nu_of_word):
    vec = CountVectorizer(ngram_range= nu_of_word, stop_words='english')
    bag_of_words = vec.fit_transform(column)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:top_nu_of_words]
    
#data=data.drop(['address','rest_type', 'type', 'menu_item', 'votes'],axis=1)
import pandas

# Randomly sample 60% of your dataframe
df_percent = data.sample(frac=0.5)
```

```python
df_percent.set_index(['Services'])
indices = pd.DataFrame(df_percent.index)
```

```python
from sklearn.metrics.pairwise import linear_kernel
# Creating tf-idf matrix
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=0, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_percent['Services'])

cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def recommend(Services, cosine_similarities = cosine_similarities):
    
    # Create a list to put top restaurants
    recommend_Services = []
    
    # Find the index of the hotel entered
    idx = indices[indices == Services].index[6]
    
    # Find the restaurants with a similar cosine-sim value and order them from bigges number
    score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
    
    # Extract top 30 restaurant indexes with a similar cosine-sim value
    top30_indexes = list(score_series.iloc[0:31].index)
    
    # Names of the top 30 restaurants
    for each in top30_indexes:
        recommend_Services.append(list(df_percent.index)[each])
    
    # Creating the new data set to show similar restaurants
    df_new = pd.DataFrame(columns=['Menu','Services', 'cost'])
    
    # Create the top 30 similar restaurants with some of their columns
    for each in recommend_Services:
        df_new = df_new.append(pd.DataFrame(df_percent[['Menu','Services', 'cost']][df_percent.index == each].sample()))
    
    # Drop the same named restaurants and sort only the top 10 by the highest rating
    df_new = df_new.drop_duplicates(subset=['Menu','Services', 'cost'], keep=False)
    df_new = df_new.sort_values(by='cost', ascending=False).head(10)
    
    print('TOP %s Services LIKE %s : ' % (str(len(df_new)), recommend_Services))
    
    return df_new
recommend('zula decoration food service water')
```
